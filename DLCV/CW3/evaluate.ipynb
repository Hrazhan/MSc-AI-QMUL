```python
def plot_train_accuraccy(epoches, dataset, alex, vgg, google, res):
    plt.rcParams['figure.figsize'] = (10.0, 8.0)  # set default size of plots
    plt.rcParams['image.interpolation'] = 'nearest'
    plt.rcParams['image.cmap'] = 'gray'
    epoch_list = range(1,epoches + 1)
    dataset = dataset
    plt.title(f"Training Accuracy ({dataset})")
    plt.plot(epoch_list, alex['train_acc'])
    plt.plot(epoch_list, vgg['train_acc'])
    plt.plot(epoch_list, google['train_acc'])
    plt.plot(epoch_list, res['train_acc'])
    plt.xlabel('epoch')
    plt.ylabel('Acc Value')
    plt.legend(['AlexNet', 'VGG', 'GoogLeNet', 'ResNet'], loc='upper left')
    plt.show()
    
def plot_test_accuraccy(epoches, dataset, alex, vgg, google, res):
    plt.rcParams['figure.figsize'] = (10.0, 8.0)  # set default size of plots
    plt.rcParams['image.interpolation'] = 'nearest'
    plt.rcParams['image.cmap'] = 'gray'
    epoch_list = range(1,epoches + 1)
    dataset = dataset
    plt.title(f"Test Accuracy ({dataset})")
    plt.plot(epoch_list, alex['test_acc'])
    plt.plot(epoch_list, vgg['test_acc'])
    plt.plot(epoch_list, google['test_acc'])
    plt.plot(epoch_list, res['test_acc'])

    plt.xlabel('epoch')
    plt.ylabel('Acc Value')
    plt.legend(['AlexNet', 'VGG', 'GoogLeNet', 'ResNet'], loc='upper left')
    plt.show()
    
def plot_train_loss(epoches, dataset, alex, vgg, google, res):
    plt.rcParams['figure.figsize'] = (10.0, 8.0)  # set default size of plots
    plt.rcParams['image.interpolation'] = 'nearest'
    plt.rcParams['image.cmap'] = 'gray'
    epoch_list = range(1,epoches + 1)
    dataset = dataset
    plt.title(f"Training Loss({dataset})")
    plt.plot(epoch_list, alex['train_loss'])
    plt.plot(epoch_list, vgg['train_loss'])
    plt.plot(epoch_list, google['train_loss'])
    plt.plot(epoch_list, res['train_loss'])
    plt.xlabel('epoch')
    plt.ylabel('Loss Value')
    plt.legend(['AlexNet', 'VGG', 'GoogLeNet', 'ResNet'], loc='upper left')
    plt.show()
    
def plot_test_loss(epoches, dataset, alex, vgg, google, res):
    plt.rcParams['figure.figsize'] = (10.0, 8.0)  # set default size of plots
    plt.rcParams['image.interpolation'] = 'nearest'
    plt.rcParams['image.cmap'] = 'gray'
    epoch_list = range(1,epoches + 1)
    dataset = dataset
    plt.title(f"Test Loss ({dataset})")
    plt.plot(epoch_list, alex['test_loss'])
    plt.plot(epoch_list, vgg['test_loss'])
    plt.plot(epoch_list, google['test_loss'])
    plt.plot(epoch_list, res['test_loss'])
    plt.xlabel('epoch')
    plt.ylabel('Loss Value')
    plt.legend(['AlexNet', 'VGG', 'GoogLeNet', 'ResNet'], loc='upper left')
    plt.show()
```
